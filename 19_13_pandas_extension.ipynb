```python
import pandas as pd
df = pd.read_csv('productstar_df.csv', index_col=0)
```

```python
import numpy as np
```

```python
df_compl = df[df['order_state'] == 'complete']
```

```python
df_compl['shipped_at'] = pd.to_datetime(df_compl['shipped_at'])
df_compl['completed_at'] = pd.to_datetime(df_compl['completed_at'])
```

```python
df_compl['day'] = df_compl['completed_at'].dt.round('D')
df_compl
```

```python
df_compl_12_18 = df_compl[(df_compl['day']>='2018-12-01') & (df_compl['day']<'2019-01-01')]
df_compl_12_19 = df_compl[(df_compl['day']>='2019-12-01') & (df_compl['day']<'2020-01-01')]
```

```python
df_compl_12_19[['item_total']].sum() / df_compl_12_18[['item_total']].sum() 
```

```python
print(df_compl_12_19[['item_total']].sum())
print(df_compl_12_18[['item_total']].sum())
```

```python
df_compl['month'] = df_compl['completed_at'].dt.month
df_compl['year'] = df_compl['completed_at'].dt.year
```

```python
df_compl
```

```python
g = df_compl.groupby(['month', 'year'], as_index=False)[['item_total']].sum()
```

```python
print(g)
```

```python
g1 = g[(g['year'] == 2018) & (g['month'] == 12)]
g2 = g[(g['year'] == 2019) & (g['month'] == 12)]
```

```python
print(g1)
print(g2)
```

```python
delta = (g2[['item_total']].sum()) / (g1[['item_total']].sum())
```

```python
del_sel = (delta - 1)*100
print(round(del_sel,0))
```
